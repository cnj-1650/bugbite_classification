# Running LIME on a Inception v3 transfer learning model

* https://github.com/marcotcr/lime/blob/master/doc/notebooks/Tutorial%20-%20Image%20Classification%20Keras.ipynb

* model_v3 is my bug bite classification model with 7 classes, transfer learning from Inception v3
    * remove classification layer of Inception v3
    * freezes all layers
    * add in Flatten, ReLU, Dropout, Softmax
    * Softmax gives the prob dis of it being the 7 classes: 5 bug bites and 2 negative (rash, normal skin)
    

```python
import os
import keras
from keras.applications import inception_v3 as inc_net
from keras.preprocessing import image
from keras.applications.imagenet_utils import decode_predictions
from skimage.io import imread
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
print('Notebook run using keras:', keras.__version__)
```

```python
import tensorflow as tf

# load in model
model_dir = '/home/eyan/Desktop/bug_bite_old/models/model_v3.h5'

model = tf.keras.models.load_model(model_dir)
```

```python
def transform_img_fn(path_list):
    out = []
    for img_path in path_list:
        img = image.load_img(img_path, target_size=(299, 299))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        x = inc_net.preprocess_input(x)
        out.append(x)
    return np.vstack(out)
```

```python
# load in the validation image data

valid_data_dir = '/home/eyan/Desktop/bug_bite/web_scraping/useful/validation'

test_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255.)

test_generator = test_datagen.flow_from_directory(
    valid_data_dir, 
    target_size=(299, 299),
    batch_size=1,shuffle=False,
    class_mode='categorical')

# getting the names of the files and the # of samples
imagenames = test_generator.filenames

nb_samples = len(imagenames)

print(nb_samples)
```

```python
path_list = []

for name in imagenames:
    path_list.append(valid_data_dir+'/'+name)

images = transform_img_fn(path_list)

plt.imshow(images[0] / 2 + 0.5)
predictions = model.predict(images)
```

```python
%load_ext autoreload
%autoreload 2
import os,sys
try:
    import lime
except:
    sys.path.append(os.path.join('..', '..')) # add the current directory
    import lime
from lime import lime_image
```

```python
explainer = lime_image.LimeImageExplainer()
```

```python
%%time
# Hide color is the color for a superpixel turned OFF. Alternatively, if it is NONE, the superpixel will be replaced by the average of its pixels
explanation = explainer.explain_instance(images[4], model.predict, top_labels=5, hide_color=0, num_samples=100)
```

# Marking boundaries and mask that make it predict for the top choice

```python
from skimage.segmentation import mark_boundaries
temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=True)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
```

```python
temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5, hide_rest=False)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
```
