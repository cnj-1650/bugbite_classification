```python
import tensorflow as tf

model = tf.keras.models.load_model('/home/eyan/Desktop/bug_bite_old/models/8_class_mixed7_79p.h5')
```

```python
import os
MODEL_DIR = '/home/eyan/Desktop/bug_bite_old/models/bug_model'
version = 1
os.environ["MODEL_DIR"] = MODEL_DIR
```

```python
export_path = os.path.join(MODEL_DIR, str(version))
print('export_path = {}\n'.format(export_path))

tf.keras.models.save_model(
    model,
    export_path,
    overwrite=True,
    include_optimizer=True,
    save_format=None,
    signatures=None,
    options=None
)

print('\nSaved model:')
!ls -l {export_path}
```

```python
# this is running in bash -- could also be run in terminal

# running in the terminal
# !tensorflow_model_server --rest_api_port=8501 --model_base_path="path-to-directory/tensorflow_serving/my_model" --model_name=bug_model

# alternatively, run with docker : https://towardsdatascience.com/deploying-deep-learning-models-using-tensorflow-serving-with-docker-and-flask-3b9a76ffbbda
```

```python
%%bash --bg 
nohup tensorflow_model_server \
  --rest_api_port=8501 \
  --model_name=bug_model \
  --model_base_path=/home/eyan/Desktop/bug_bite_old/models/bug_model/ >server.log 2>&1
```
