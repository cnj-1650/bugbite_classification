```python
import numpy as np
import keras
from keras.applications.imagenet_utils import decode_predictions
import skimage.io 
import skimage.segmentation
import copy
import sklearn
import sklearn.metrics
from sklearn.linear_model import LinearRegression
import warnings

# taking in an image
img_path = '/home/eyan/Desktop/bug_bite/Flask_app/bugbiteID/static/brown_recluse_spiderimage81.jpg'


Xi = skimage.io.imread(img_path)
Xi = skimage.transform.resize(Xi, (299,299)) 

# what is this???? Do I need to do this normalization??
Xi = (Xi - 0.5)*2 #Inception pre-processing
skimage.io.imshow(Xi/2+0.5)
```

```python
# making predictions - need to expand a newaxis for Xi because it needs to be in (,299,299,3)
predict = model.predict(Xi[np.newaxis,:,:,:])
```

```python
superpixels = skimage.segmentation.quickshift(Xi, kernel_size=4,max_dist=200, ratio=0.2)
num_superpixels = np.unique(superpixels).shape[0]
num_superpixels
```

```python
skimage.io.imshow(skimage.segmentation.mark_boundaries(Xi/2+0.5, superpixels))
```

```python
# higher perturbation would make more reliable explanation
num_perturb = 150

# a random generator of 0 and 1
perturbations = np.random.binomial(1, 0.5, size=(num_perturb, num_superpixels))
perturbations[0] #Show example of perturbation
```

```python
# a function to apply the perturbation vector to the image to perturbe it, segment is the superpixels

def perturb_image(img,perturbation,segments):
  active_pixels = np.where(perturbation == 1)[0]
  mask = np.zeros(segments.shape)
  for active in active_pixels:
      mask[segments == active] = 1 
  perturbed_image = copy.deepcopy(img)
  perturbed_image = perturbed_image*mask[:,:,np.newaxis]
  return perturbed_image
```

```python
skimage.io.imshow(perturb_image(Xi/2+0.5,perturbations[0],superpixels))
```

# Now onto the heavier duty mass sample for spiders

```python
# going to do it for all the perturbations generated from the binomial
predictions = []
for pert in perturbations:
  perturbed_img = perturb_image(Xi,pert,superpixels)
  pred = model.predict(perturbed_img[np.newaxis,:,:,:])
  predictions.append(pred)

# making predictions an numpy array
predictions = np.array(predictions)
predictions.shape
```

```python
# Perturbation with all superpixels enabled would be the original image
original_image = np.ones(num_superpixels)[np.newaxis,:] 
distances = sklearn.metrics.pairwise_distances(perturbations,original_image, metric='cosine').ravel()
distances.shape
```

```python
# using kernel to computer weight
kernel_width = 0.25
weights = np.sqrt(np.exp(-(distances**2)/kernel_width**2)) #Kernel function
weights.shape
```

```python
top_pred_class = np.argmax(predict)
print(top_pred_class)
```

```python
class_to_explain = top_pred_class
simpler_model = LinearRegression()
simpler_model.fit(X=perturbations, y=predictions[:,:,class_to_explain], sample_weight=weights)
coeff = simpler_model.coef_[0]
coeff
```

```python
# computer the top feature (the super pixels)

num_top_features = 4
top_features = np.argsort(coeff)[-num_top_features:] 
top_features
```

```python
# the top feature used in the image - the super pixels used
mask = np.zeros(num_superpixels) 
mask[top_features]= True #Activate top superpixels
skimage.io.imshow(perturb_image(Xi/2+0.5,mask,superpixels) )
```
