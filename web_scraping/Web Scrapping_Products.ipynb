# Defininig a function to do everything at once

```python
# defining the function to take URL and category and output (returns) a dataframe
def df_products(URL, category):
    # get the html of the page
    page = requests.get(URL)
    soup = BeautifulSoup(page.text, 'html.parser')

    links = []
    images = []
    titles = []
    prices = []

    for link in soup.find_all('a',class_="search-result-productimage gridview display-block"):
        links.append(link['href'])

    for item in soup.find_all('div', class_="search-result-gridview-item clearfix arrange-fill"):
        image_src = item.img['src']
        images.append(image_src)

    for product in soup.find_all('a',class_="product-title-link line-clamp line-clamp-2 truncate-title"):
        title = product.span
        titles.append(title)

    for price in soup.find_all('span', class_="price-main-block"):
        x = price.find('span',class_="visuallyhidden").text
        prices.append(x)
    
    ts = []
    for t in titles:
        t = re.sub('<[^<]+?>', '', str(t))
        ts.append(t)
    
    print("links",len(links))
    print("images",len(images))
    print("titles",len(titles))
    print("prices",len(prices))
    

    dataframe = {'title':ts,'price':prices,'image':images,'link':links}
    column_names = ['title','price','image','link']
    db = pd.DataFrame(dataframe, columns = column_names)
    db['link'] = 'https://www.walmart.com'+ db['link']
    db['category'] = category
    return db
```

```python
URL = 'https://www.walmart.com/search/?query=calamine%20lotion'
db = df_products(URL,'calamine lotion')
# save db to csv
db.to_csv('./webpage_resources/calamine_lotion.csv',index=False)
```

# Import and concat all the csv files

```python
# to get the list of the files
!ls webpage_resources -1
```

```python
# file name
file_name = ['aloe_vera.csv'
,'antibiotic_ointment.csv'
,'anti_itch_cream.csv'
,'benadryl.csv'
,'calamine_lotion.csv'
,'witch_hazel.csv']
```

```python
frames = []

# for loop for go through the list and add to frames
for file in file_name:
    src = './webpage_resources/'+file
    # print(src)
    df = pd.read_csv(src)
    frames.append(df)

# concat all dataframes in frames into one final result
result = pd.concat(frames)
```

```python
len(result)
```

# Make SQLite database

```python
from sqlalchemy import create_engine

# save a sqlite file called bugbite
engine = create_engine('sqlite:///bugbite.sqlite', echo=False)
```

```python
# convert dataframe to sql, labeling the index as id
result.to_sql('products',con=engine,if_exists='replace',index_label='id')
```
