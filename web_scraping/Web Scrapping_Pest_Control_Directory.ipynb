# Web Scrapping for pest control company

```python
import requests

# the url to grab from
URL = 'https://pest-control.regionaldirectory.us'

# get the html of the page
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup

# parse
soup = BeautifulSoup(page.text, 'html.parser')


import re

# getting the name of the state to use to append to the main url
states=[]
up_states = []
for item in soup.find_all('div', attrs={'class': 'o'}):
    for row in item.find_all('a'):
        m = re.search('(?<=>)(.*?)(?=<span)',str(row))
        up_states.append(m[0])
        m = re.sub(r"\s+", '-', m[0])
        states.append(m.lower())

states[-1] = 'washington-dc'
print(states)
```

```python
print(up_states)
```

# For all states

```python
import pandas as pd
column_names = ['name','website','phone','city','county','state']
db = pd.DataFrame(columns = column_names)
```

```python
for s in states:
    state_url = URL+"/"+s+".htm"
    state_page = requests.get(state_url)

    state_soup = BeautifulSoup(state_page.text, 'html.parser')

    table = state_soup.table

    for item in table.find_all('td', attrs={'valign': 'top','width':'360'}):
        #print(item)
        a = item.a
        if a != None:
            name = a.string
            city = a.next_sibling.next_sibling.strip()
            phone = a.next_sibling.next_sibling.next_sibling.next_sibling.strip()
        web = item.find('div',attrs={'class':'d'})
        if web != None:
            website = web.string
        county = item.find('i')
        if county !=None:
            county = county.string
            state = s
        #print(item.a)
        #print('/n')
        #print(item)
        #n = re.search('(?<=>)(.*?)(?=</a><br)',s)
            db = db.append({'name':name,'website':website,'phone':phone,'city':city,'county':county,'state':state},ignore_index=True)  
```
